Program ini dibuat untuk melakukan stock management di perusahaan supermarket (menambahkan barang, melihat seluruh inventory).


Program inventory management dibawah dapat melakukan
1. input data
2. print data

Contoh Input: 1 <br>

pilih menu atau keluar  <br>
masukkan nama barang: SoKlin <br>
masukkan kategori barang: Baju <br>
masukkan harga (rupiah): 10000 <br> <br>

Output:  <br>
Data berhasil diinputkan <br>
nama barang: SoKlin <br>
jenis barang: Baju <br>
harga barang: 10000 <br>

Contoh Input: 2

Output: 
Daftar semua barang <br> <br>
No | Nama Barang | Jenis Barang | Harga Barang <br>
1 | SoKlin | Baju | 10000


```python
# define variable
data = []

# function 1
def inputBarang(nama, kategori, harga):
    barang = [nama, kategori, harga]
    data.append(barang)

# test
# inputBarang("apel", "buah", 20000)
# inputBarang("jeruk", "buah", 50000)

# function 2
def showBarang():
    if len(data) == 0:
        print("Data belum ada")
    else:
        # menampilkan seluruh barang
        print("No.|Nama|Kategori|Harga")
        counter = 1
        for item in data:
            # cara 1
            # print(counter, "|", item[0], "|", item[1], "|", item[2])

            # cara 2
            print(f"{counter} | {item[0]} | {item[1]} | {item[2]}")
            
            # increment counter
            counter += 1

# test
# showBarang()
```

```python
# program
isRunning = True

while isRunning == True:
    perintah = input("pilih menu atau exit")
    if perintah == "exit":
        # cara break
        break

        # cara status
        # isRunning = False
    elif perintah == "1":
        nama = input("masukkan nama")
        kategori = input("masukkan kategori")
        harga = input("masukkan harga")
        inputBarang(nama, kategori, harga)

        print("terima kasih")
    elif perintah == "2":
        showBarang()
    else: # optional -> handle menu diluar 1,2,exit
        print("menu invalid")
```
